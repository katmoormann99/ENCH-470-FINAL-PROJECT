```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import nan as Nan
from IPython.display import display
```

```python
# Downloaded CSV from internet 

# https://aqs.epa.gov/aqsweb/airdata/download_files.html

df_original = pd.read_csv('annual_aqi_by_county_2020.csv')
```

```python
# Clean Up CSV 

# Drop unwanted column names

df= df_original.drop(columns=['Days with AQI', 'Good Days', 'Moderate Days', 'Unhealthy for Sensitive Groups Days','Unhealthy Days'])
df= df.drop(columns=['Very Unhealthy Days','Hazardous Days','Days CO','Days NO2','Days Ozone','Days SO2','Days PM2.5','Days PM10'])

# Drop the word county

pdToList_county = list(df['County'])
def first_word(pdToList_county):
    my_list = [x.split()[0] for x in pdToList_county]
    return my_list

County = first_word(pdToList_county)

df.insert(2, 'First Word of Each County', County)

# Creating abbreivation through Excel 

df.to_csv('df.csv')
df = pd.read_csv('df_with_abbreivations.csv')  

```

```python
# https://www.ers.usda.gov/data-products/county-level-data-sets/

edu_original = pd.read_csv('Education (1).csv',encoding='ISO-8859-1')
```

```python
# Cleaning Up CSV
neglect2013 = [col for col in edu_original if '2013' in col]
edu = edu_original.drop(columns=neglect2013)

neglect2003 = [col for col in edu_original if '2003' in col]
edu = edu.drop(columns=neglect2003)
edu

neglect1970 = [col for col in edu if '1970' in col]
edu = edu.drop(columns=neglect1970)
edu

neglect1980 = [col for col in edu if '1980' in col]
edu = edu.drop(columns=neglect1980)
edu

neglect1990 = [col for col in edu if '1990' in col]
edu = edu.drop(columns=neglect1990)
edu

neglect2000 = [col for col in edu if '2000' in col]
edu = edu.drop(columns=neglect2000)
edu

edu= edu.drop(columns=['FIPS Code'])
edu= edu.drop(columns=["Less than a high school diploma, 2015-19","High school diploma only, 2015-19","Some college or associate's degree, 2015-19"])
edu= edu.drop(columns=["Bachelor's degree or higher, 2015-19","Percent of adults with less than a high school diploma, 2015-19"])
edu= edu.drop(columns=["Percent of adults completing some college or associate's degree, 2015-19"])


pdToList_area_name = list(edu['Area name'])

pdToList_area_name

def first_word(pdToList_area_name):
    my_list = [x.split()[0] for x in pdToList_area_name]
    return my_list

Area_name = first_word(pdToList_area_name)

edu.insert(2, 'First Word of Each County', Area_name)

edu.to_csv('edu.csv')

# Add needed abbreivations via Excel 
edu=pd.read_csv('edu_with_abbreivations.csv')
```

```python
# First dataset include an inner join between AQI data and Education Data 

Dataset = pd.merge(left=df, right=edu, left_on=['Abbreviation'], right_on=['Abbreviation'])


Date_Set_1 = Dataset.drop(columns="Area name")
Date_Set_1= Date_Set_1.drop(columns="State_y")
Date_Set_1= Date_Set_1.drop(columns="First Word of Each County_x")
Date_Set_1= Date_Set_1.drop(columns="First Word of Each County_y")
Date_Set_1=Date_Set_1.rename(columns={"State_x": "State"})
Date_Set_1.to_csv('Data_Set_1.csv')


```

```python
# https://www.ers.usda.gov/data-products/county-level-data-sets/

empl_original = pd.read_csv('Unemployment.csv')

# Clean Up CSV
empl=empl_original[~empl_original.Attribute.str.contains("2000")]
empl=empl[~empl.Attribute.str.contains("2001")]
empl=empl[~empl.Attribute.str.contains("2002")]
empl=empl[~empl.Attribute.str.contains("2003")]
empl=empl[~empl.Attribute.str.contains("2004")]
empl=empl[~empl.Attribute.str.contains("2005")]
empl=empl[~empl.Attribute.str.contains("2006")]
empl=empl[~empl.Attribute.str.contains("2007")]
empl=empl[~empl.Attribute.str.contains("2008")]
empl=empl[~empl.Attribute.str.contains("2009")]
empl=empl[~empl.Attribute.str.contains("2010")]
empl=empl[~empl.Attribute.str.contains("2011")]
empl=empl[~empl.Attribute.str.contains("2012")]
empl=empl[~empl.Attribute.str.contains("2013")]
empl=empl[~empl.Attribute.str.contains("2014")]
empl=empl[~empl.Attribute.str.contains("2015")]
empl=empl[~empl.Attribute.str.contains("2016")]
empl=empl[~empl.Attribute.str.contains("2017")]
empl=empl[~empl.Attribute.str.contains("2019")]
empl=empl[~empl.Attribute.str.contains("2018")]
empl=empl[~empl.Attribute.str.contains("Civilian")]
empl=empl[~empl.Attribute.str.contains("rate")]

# Create one database containing data for the Employed
employed = empl[~empl.Attribute.str.contains("Unemployed")]
employed = employed.rename(columns={"Value": "# Employed"})
employed = employed.drop(columns=['Attribute'])

# Create a second database containing data for the Unemployed

unemployed = empl[~empl.Attribute.str.contains("Employed")]
unemployed = unemployed.rename(columns={"Value": "# Unemployed"})
unemployed=unemployed.drop(columns=['Attribute'])


# Merge Together both the employed and unemployed data frames
Employment_df = pd.merge(left=employed, right=unemployed, left_on=['Area_name'], right_on=['Area_name'])
pdToList_area_name = list(Employment_df['Area_name'])

# Creating a column based on the First Word of Each County
def first_word(pdToList_area_name):
    my_list = [x.split()[0] for x in pdToList_area_name]
    return my_list

Area_name = first_word(pdToList_area_name)

Employment_df.insert(2, 'First Word of Each County', Area_name)
Employment_df.to_csv('Employment_df.csv')

# Add abbreviations within excel 
Employment_numbers=pd.read_csv('Employment_df_with_abbreivations.csv')
```

```python
# Create second data base containing information from database 1 and the new employment numbers for each county

Dataset = pd.merge(left=Date_Set_1, right=Employment_numbers, left_on=['Abbreviation'], right_on=['Abbreivation'])
Data_Set_2=Dataset.drop_duplicates(subset=["# Employed","# Unemployed"])
Data_Set_2.to_csv('Data_Set_2.csv')

# Clean up CSV over excel
updated_12_02 = pd.read_csv('Data_Set_2_cleanedUP.csv')
```
